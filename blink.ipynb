```python
import cv2
import numpy as np
import dlib
import pyglet
from math import hypot
```

```python
music = pyglet.resource.media('alarm.mp3')
```

```python
cap = cv2.VideoCapture(0)

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")


def midpoint(p1 ,p2):
    return int((p1.x + p2.x)/2), int((p1.y + p2.y)/2)


count = 0

while True:
    _, frame = cap.read()
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(frame)
    
    for face in faces:
        landmarks = predictor(gray,face)
        
        # points of out left and rightmost parts of eye
        x = landmarks.part(36).x
        y = landmarks.part(36).y
        
        left_point = (x,y)
        
        x1 = landmarks.part(39).x
        y1 = landmarks.part(39).y
        
        right_point = (x1,y1)
        
        #draw horizontal line in left eye
        hor_line = cv2.line(frame, left_point, right_point, (0, 255, 0), 2)
        
        
        # points of out left and rightmost parts of eye
        x = landmarks.part(36).x
        x_r = landmarks.part(42).x
        y_r = landmarks.part(42).y
        
        left_point_r = (x_r,y_r)
        
        x_r_1 = landmarks.part(45).x
        y_r_1 = landmarks.part(45).y
        
        right_point_r = (x_r_1,y_r_1)
        
        hor_line = cv2.line(frame, left_point_r, right_point_r, (0, 255, 0), 2)
        
        top_l = midpoint(landmarks.part(37),landmarks.part(38))
        bottom_l = midpoint(landmarks.part(40),landmarks.part(41))
        
        
        # draw vertical line
        ver_line_l = cv2.line(frame, top_l, bottom_l, (0, 255, 0), 2)
        
        top_r = midpoint(landmarks.part(43),landmarks.part(44))
        bottom_r = midpoint(landmarks.part(47),landmarks.part(46))
        
        # draw vertical line
        ver_line_l = cv2.line(frame, top_r, bottom_r, (0, 255, 0), 2)
        
        
        # length between eyelids (left eye)
        vl_length = hypot((top_l[0] - bottom_l[0]),(top_l[1] - bottom_l[1]))
        
        # length between eyelids (right eye)
        
        vl_length_r = hypot((top_r[0] - bottom_r[0]),(top_r[1] - bottom_r[1]))
        
        
        if vl_length < 8 and vl_length_r < 8:
            count += 1
            
            if(count > 25):
                music.play()
                pyglet.app.run()
                
        else:
            print(vl_length)
        
        

        #cv2.circle(frame, (x, y), 3,(7, 4, 255) ,2)
        #cv2.circle(frame, (x1, y1), 3,(7, 4, 255) ,2)
    
    cv2.imshow("Frame" , frame)
    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
